```python
df.reset_index(drop=True, inplace=True)

# Function to check if a value is a string and try converting to datetime
def convert_to_date(value):
    if isinstance(value, str):  # Check if the value is a string
        try:
            # Convert string to datetime and then format as MM/DD/YYYY
            return pd.to_datetime(value).strftime('%-m/%-d/%Y')  # For Unix/Mac
            # return pd.to_datetime(value).strftime('%#m/%#d/%Y')  # For Windows
        except ValueError:
            return value  # If conversion fails, return the original value
    return value

# Apply the conversion function to each row in the 'SaleDate' column
for i in range(len(df['SaleDate'])):
    df.at[i, 'SaleDate'] = convert_to_date(df.at[i, 'SaleDate'])
```

# handell missing value 

```python
df.info()
```

```python
df['SaleDate'] = df['SaleDate'].apply(convert_to_date)
```

```python
df.info()
```

```python
df['SaleDate'] = pd.to_datetime(df['SaleDate'], errors='coerce')
```

```python
df['PropertyAddress'] = df['PropertyAddress'].astype(str)
df['LandUse'] = df['LandUse'].astype(str)

```
